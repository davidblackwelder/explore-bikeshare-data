## Clean

```r
# Select variables
ny %>%
  select(Trip.Duration, Start.Station, End.Station)
```

```r
# Select variables using index
ny %>%
  select(4:6)
```

```r
ny %>%
  select(ends_with("Station"))
```

```r
# Changing variable order
ny %>%
  select(Start.Station, End.Station, Trip.Duration, everything())
```

```r
# Changing variable name
ny %>%
  rename("duration" = "Trip.Duration") %>%
  head()
```

```r
head(ny)
```

```r
# Filter rows
ny %>%
  select(Start.Station, End.Station, Trip.Duration, User.Type) %>%
  filter(Trip.Duration > 800 & 
         User.Type != "Subscriber")
```

```r
# Recode data
ny %>%
  select(Gender, everything()) %>%
  mutate(Gender = recode(Gender,
                         "Male" = "man",
                         "Female" = "woman"))
```

```r
# Dealing with missing data
# mean(ny$Birth.Year) # Returns N/A because it can't be calculated due to missing values
mean(ny$Birth.Year, na.rm = TRUE) # na is for missing values, .rm is to remove and so na.rm = TRUE means remove na values.
```

```r
# Dealing with duplicates
ny %>%
  distinct()
```

## Manipulate

```r
# Create or change a variable (mutate)
ny %>%
  mutate(trip_in_minutes = Trip.Duration/60) %>%
  select(Start.Station, End.Station, Trip.Duration, trip_in_minutes) %>%
  head()
```

```r
# Conditional change (if_else)
ny %>%
  mutate(age = (2024 - Birth.Year)) %>%
  mutate(trip_in_minutes = Trip.Duration/60) %>%
  select(Start.Station, End.Station, Trip.Duration, trip_in_minutes, Birth.Year, age) %>%
  mutate(age_desc = 
           if_else(age > 17,
                   "adult",
                   "kid"))
```

## Describe and summarize

```r
# Range / spread
min(ny$Birth.Year, na.rm = TRUE)
max(ny$Birth.Year, na.rm = TRUE)
range(ny$Birth.Year, na.rm = TRUE)
IQR(ny$Birth.Year, na.rm = TRUE)
```

```r
# Centrality
mean(ny$Birth.Year, na.rm = TRUE)
median(ny$Birth.Year, na.rm = TRUE)
```

```r
# Variance
var(ny$Birth.Year, na.rm = TRUE)
```

```r
summary(ny$Birth.Year)
```

```r
ny %>%
  drop_na(User.Type) %>%
  group_by(User.Type) %>%
  summarize(Lower = min(Trip.Duration)/60,
            Average = mean(Trip.Duration, na.rm = TRUE)/60,
            Upper = max(Trip.Duration)/60,
            Difference = (max(Trip.Duration)/60) - (min(Trip.Duration)/60)) %>%
  arrange(Average)
```
